```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

$f(x)=\frac{e^x-e^-x}{e^x+e^-x}$

```python
x = np.linspace(-5, 5)
e = np.e
```

```python
def f(x, e):
    return ((e ** x)-(e ** -x))/((e ** x)+(e ** -x))
```

```python
plt.plot(x, f(x, e), 'c')
plt.xlabel("Valor x")
plt.ylabel("Funcion f(x)")
plt.grid(True)
plt.title("GRAFICA TANGENTE HIPERBOLICA", fontsize = 20)
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def escalon(x, c):
    return np.piecewise(x, [x < 0, x > 0], [0, 1])
```

```python
c = np.linspace(0, 5)
```

```python
plt.plot(x, escalon(x, c), 'm')
plt.axis([x[0], x[-1], 0, 2.5])
plt.grid(True)
plt.xlabel("Valor x")
plt.ylabel("Funcion f(x)")
plt.title("GRAFICA ESCALON", fontsize = 20)
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
sigma=0.01
x = np.linspace(-5,5)
e = np.e
pi = np.pi
x = np.linspace(-50, 50)
```

```python
def f(x):
    return (1/np.sqrt(2*pi*sigma**2))*(e**-((x**2)/2*(sigma**2)))
```

```python
plt.plot(x, f(x), 'r')
plt.grid(True)
plt.xlabel("Valor x")
plt.ylabel("Funcion f(x)")
plt.title("GRAFICA SIGMOIDAL", fontsize = 20)
```
